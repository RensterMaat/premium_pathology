```python
from pathlib import Path

r = Path('/mnt/hpc/rens/CLAM/scripts')
log = r / 'slurm-20616639.out'

with open(log, 'r') as f:
    text = f.readlines()
```

```python
fold_ixs = [ix for (ix, line) in enumerate(text) if 'fold' in line.lower()]
```

```python
import matplotlib.pyplot as plt

for fold in range(6):
    fold_text = text[fold_ixs[fold]:fold_ixs[fold+1]]
    completion_ixs = [ix for (ix, line) in enumerate(fold_text) if 'epoch' in line.lower()]

    train_loss, val_loss = [],[]
    for ix in completion_ixs:

        train_loss.append(float(fold_text[ix].split()[3][:-1]))
        val_loss.append(float(fold_text[ix+4].split()[3][:-1]))

    plt.figure()
    plt.title(f'Fold {fold}')
    plt.plot(range(len(train_loss)), train_loss, label='train loss')
    plt.plot(range(len(val_loss)), val_loss, label='val loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.show()
```
