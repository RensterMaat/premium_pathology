```python
import json
from pathlib import Path

r = Path(r'C:\Users\Rens\Data\pathology\annotations')

with open(r / '2021-08-12 13.48.40.json') as f:
    data = json.load(f)
```

```python
geo_data=[]
for id, values in data['nuc'].items():
	dict_data = {}
	cc = values['contour']
	cc.append(cc[0])

	# dict_data={}
	# cc=data[i][1]['contour']
	# cc.append(cc[0])  
	dict_data["type"]="Feature"
	dict_data["id"]="PathCellObject"
	dict_data["geometry"]={
		"type":"Polygon",
		"coordinates":[cc]
	}
	dict_data["properties"]={
		"isLocked":"false",
		"measurements":[],
		"classification": {
			"name": values['type']
		}
	}
	
	geo_data.append(dict_data)
```

```python
import geojson

with open(r / 'geojson.json', 'w') as file:
    geojson.dump(geo_data, file)
```
